```python
cat_df=df.select_dtypes(include='object')
num_df=df.select_dtypes(exclude='object')

num_df.columns
```

```python
# Filling the null values in categorical columns with mode and the numerical columns with mean.
for i in df.columns:
    if i in cat_df:
        df[i].fillna(df[i].mode()[0],inplace = True)
    elif i in num_df:
        df[i].fillna(df[i].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
duplicates = df[df.duplicated(keep=False)]
duplicates
```

```python
df['GENDER'].value_counts()
```

```python
df['TARGET'] = df['TARGET'].replace({0:'Payers', 1:'Defaulters'})
df['GENDER'] = df['GENDER'].replace('XNA',df['GENDER'].mode()[0])
df['NAME_FAMILY_STATUS'] = df['NAME_FAMILY_STATUS'].replace('Civil marriage','Married')
```

```python
def cate_occupation(occupation):
    if occupation in ['Laborers', 'Drivers','Cleaning staff', 'Cooking staff', 'Low-skill Laborers',
                      'Waiters/barmen staff','Security staff']:
        return 'Unskilled'
    else:
        return 'Skilled'

df['Skill_Level'] = df['OCCUPATION_TYPE'].apply(cate_occupation)
```

```python
df['NAME_INCOME_TYPE'] = df['NAME_INCOME_TYPE'].replace({'State servant':'Working', 'Commercial associate':'Working'})
df['NAME_INCOME_TYPE'] = df['NAME_INCOME_TYPE'].replace({'Student':'Unemployed','Businessman':'Working','Maternity leave':'Working'})
```

```python
df['NAME_TYPE_SUITE'] = df['NAME_TYPE_SUITE'].replace({'Family':'accompanied','Spouse, partner':'accompanied','Children':'accompanied','Other_A':'accompanied','Other_B':'accompanied','Group of people':'accompanied' })
```

```python
df['NAME_EDUCATION_TYPE'] = df['NAME_EDUCATION_TYPE'].replace({'Academic degree': 'Higher education','Incomplete higher': 'Higher education'})
```

```python
df['NAME_FAMILY_STATUS'] = df['NAME_FAMILY_STATUS'].replace({'Unknown':'Married'})
```

```python
df['TOTAL_DOC_SUBMITTED'] = df['TOTAL_DOC_SUBMITTED'].replace({0: '<=1',1: '<=1',2: '>=2',3: '>=2',4: '>=2'})
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,4))
sb.boxplot(x = df['CNT_CHILDREN'], ax= ax[0])
ax[0].set_title("Boxplot of count of children")
sb.boxplot(x = df['CNT_FAM_MEMBERS'], ax= ax[1])
ax[1].set_title("Boxplot of count of family members")
plt.show()

# So based on the graphs filter the data 
df = df[(df['CNT_CHILDREN']<=2) & (df['CNT_FAM_MEMBERS']<=4)]
```

```python
def analyse(column,i,j):
    fig, axs = plt.subplots(1,3,figsize=(20,5))

    sb.countplot(x = df[column],ax = axs[0])
    axs[0].set_title(f"Count of people by {column}")


    table1 = df[df[column] == i]
    index = table1['TARGET'].value_counts().index
    values = table1['TARGET'].value_counts().values
    axs[1].pie(x = values, labels = index, autopct = '%0.2f%%' )
    axs[1].set_title(i)

    table2 = df[df[column] == j]
    index = table2['TARGET'].value_counts().index
    values = table2['TARGET'].value_counts().values
    axs[2].pie(x = values, labels = index, autopct = '%0.2f%%' )
    axs[2].set_title(j)

    for a in axs:
        a.set_xticklabels(a.get_xticklabels(), rotation=90)
        
    plt.show()
```

```python
analyse('NAME_CONTRACT_TYPE','Cash loans','Revolving loans')
print("People are taking more cash loans than revolving loans.")
print("People who take revolving loans have less defaulters.")
```

```python
analyse('GENDER','M','F')
print("Females take more loans.")
print("Males have more percentage of defaulters.")
```

```python
analyse('Car','Y','N')
print("People who donot own a car are taking more loans.")
print("People who own/donot own a car have same percentage of defaulters.")
```

```python
analyse('House','Y','N')
print("People who own a house are taking more loans.")
print("People who own/donot own a house have same percentage of defaulters.")
```

```python
# multivariant analysis
for i in cat_df:
    fig, ax = plt.subplots(figsize=(15, 5))
    sb.barplot(x='TARGET', y='AMT_INCOME_TOTAL', hue=i, data=df, ax=ax, palette="Set1", dodge=True, alpha=0.7)
    sb.barplot(x='TARGET', y='AMT_CREDIT', hue=i, data=df, ax=ax, palette="Set2", dodge=True, alpha=0.7)
    ax.set_xlabel('Target')
    ax.set_ylabel('Amount')
    ax.set_title('Comparison of AMT_INCOME_TOTAL and AMT_CREDIT by Target and income type')
    plt.show()

print("Pensioners are taking less credit compared to others. So there are few number of defaulters")
```
